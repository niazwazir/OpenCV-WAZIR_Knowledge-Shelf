```
import cv2
import matplotlib.pyplot as plt
```

```
img = cv2.imread("TestImage/butterfly_GT.bmp")
img_cvt=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2.imwrite("NEWIMAGE.png",img_cvt)
img1 = cv2.imread("NEWIMAGE.png")
cv2.imwrite("SECONDNEWIMAGE.png",img1)
img2 = cv2.imread("SECONDNEWIMAGE.png")
```

```
# Result visualization
fig = plt.figure(figsize=(18, 16), dpi= 80)


ax = plt.subplot("131")
ax.imshow(img_cvt)
ax.set_title("ORIGINAL IMAGEGT")
plt.axis('off')

ax = plt.subplot("132")
ax.imshow(img1)
ax.set_title("NEW CREATED IMAGE)")
plt.axis('off')


ax = plt.subplot("133")
ax.imshow(img2)
ax.set_title("SECOND VERSION CREATED IMAGE")
plt.axis('off')

plt.show()
```
